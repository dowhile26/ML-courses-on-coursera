<br>Мат ожидание (среднее) Парето <tex>$E[X]={{kx_m}\over{k-1}}$</tex> при k>1, в scipy по умолчанию <tex>$x_m=1$</tex>, k - параметр кривизны
<br>Дисперсия Парето <tex>$E[D]={{E[x]^2k}\over{k-2}}$</tex> при к>3
<br>Выберем параметр k=3., чтобы можно было выслить как мат ожидание, так и дисперсию

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math

k = 3. # коэффициент кривизны
xm = 1 
Ex = xm*k/(k-1) # мат ожидание Парето
Ed = (Ex**2)*k/(k-2) # Дисперсия Парето

print Ex, Ed
```

```python
pareto = sts.pareto(k)
sample = pareto.rvs(1000) # генерируем выборку объема 1000
x = np.linspace(1, 10, 100)

plt.plot(x, pareto.pdf(x), lw = 5, alpha = 0.8, label='pareto pdf') # график плотности
plt.hist(sample, bins = 50, normed=True, alpha = 0.5) # гистограмма среднего выборочного
plt.legend()
plt.show()
```

```python
i = 0
n = 5
sample5n = np.array([])
# генерируем 1000 выборок объема 5
while i < 1000:
    sample5 = pareto.rvs(n) # генерируем выборку объема 5
    mean5 = np.sum(sample5)/n # вычисляем среднее выборочное
    sample5n = np.append(sample5n, mean5) # записываем среднее выборочное каждой выборки в массив
    i += 1
    
Dn5 = Ed/n # рассчитываем дисперсию 
sigma5 = math.sqrt(Dn5) # и сигму для нормального распределения
norm5 = sts.norm(Ex, sigma5) # нормальное распределение для Парето
plt.plot(x, norm5.pdf(x), lw=5, alpha = 0.8, label='pareto pdf') # функция плотности
plt.hist(sample5n, bins = 20, normed = True, alpha = 0.5) # гистограмма среднего выборочного
plt.legend()
plt.show()
    
```

```python
i = 0
n = 10
sample10n = np.array([])
# генерируем 1000 выборок объема 10, все остальные действия аналогичны предыдущему юлоку кода
while i < 1000:
    sample10 = pareto.rvs(n)
    mean10 = np.sum(sample10)/n
    sample10n = np.append(sample10n, mean10)
    i += 1
    
Dn10 = Ed/n
sigma10 = math.sqrt(Dn10)
norm10 = sts.norm(Ex, sigma10)

plt.plot(x, norm10.pdf(x), lw=5, alpha = 0.8, label='pareto pdf')
plt.hist(sample10n, bins = 15, normed = True, alpha = 0.5)
plt.legend()
plt.show()
```
